###Importando uma nova base de dados

````python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
````

````python
deaths.head()
````

###Informações sobre as novas bases de dados

````python
print('Informações sobre a base de dados (deaths):')
print('Nº de linhas: {}'.format(deaths.shape[0]))
print('Nº de colunas: {}'.format(deaths.shape[1]))
````

###Criando uma função que retorna o último dia de dados coletados, para cada base de dados

````python
def latest_by_country(data):
  return data.groupby(['Country/Region']).sum().iloc[:,-1:]
````

####Testando a função para a base de dados 'confirmed'

````python
latest_by_country(confirmed)
````

####Testando a função para a base de dados 'recovered'

####Testando a função para a base de dados 'deaths'

````python
latest_by_country(deaths)
````

###Concatenando as três bases de dados (somente último dia)

````python
informations = [latest_by_country(confirmed), latest_by_country(deaths)]
combined = pd.concat(informations, axis = 1)
combined.columns = ['confirmed','deaths']
combined
````

###Calculando a taxa de letalidade para todas as amostras

````python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed']

print('O 1º modo de cálculo é a taxa de mortes / confirmados. ',
      (letality_rate_1*100).round(3),
      '% dos casos confirmados (nessa base de dados), vieram a óbito.')
````

###Calculando a letalidade por país

````python
letality_rate_1 = ((combined['deaths'] / combined['confirmed']) * 100).round(3)
combined['letality_rate_1'] = letality_rate_1
combined
````
